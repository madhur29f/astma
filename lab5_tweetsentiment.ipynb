```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from sklearn.metrics import classification_report, accuracy_score

def analyze_twitter_data():
    analyzer = SentimentIntensityAnalyzer()

    print("=" * 70)
    print("Loading Twitter Dataset...")
    print("=" * 70)

    try:
        df = pd.read_csv("data/lab5_Twitter_Data_sentiment.csv", encoding="latin-1").dropna()
    except FileNotFoundError:
        print("Error: Dataset file not found.")
        return

    # Label Mapping
    category_map = {
        -1.0: "Negative",
         0.0: "Neutral",
         1.0: "Positive"
    }

    df["true_label"] = df["category"].map(category_map)

    # VADER Prediction Function
    def get_vader_label(text):
        score = analyzer.polarity_scores(str(text).lower())["compound"]

        if score >= 0.05:
            return "Positive"
        elif score <= -0.05:
            return "Negative"
        else:
            return "Neutral"

    print("Applying VADER Sentiment Analysis...")
    df["vader_prediction"] = df["text"].apply(get_vader_label)

    valid_df = df.dropna(subset=["true_label", "vader_prediction"])

    # Sample Results
    print("\n" + "=" * 70)
    print("Sample Predictions")
    print("=" * 70)

    for i, row in valid_df.head(5).iterrows():
        print(f"\nTweet {i+1}:")
        print(f"Text       : {str(row['text'])[:120]}...")
        print(f"Predicted  : {row['vader_prediction']}")

    # Accuracy
    acc = accuracy_score(valid_df["true_label"], valid_df["vader_prediction"])

    print("\n" + "=" * 70)
    print("Model Performance")
    print("=" * 70)
    print(f"Accuracy : {acc:.4f}")

    # Classification Report
    print("\nClassification Report:")
    print(classification_report(valid_df["true_label"], valid_df["vader_prediction"]))

if __name__ == "__main__":
    analyze_twitter_data()
```
