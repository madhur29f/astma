```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords, wordnet
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag

# Download required resources
nltk.download("punkt")
nltk.download("stopwords")
nltk.download("averaged_perceptron_tagger")
nltk.download("wordnet")

text = "I love apples"

print("=" * 60)
print("CORRECT NLP PREPROCESSING PIPELINE")
print("=" * 60)

# Step 1: Original Input
print("\n1. Original Input Text:")
print(text)

# Step 2: Tokenization
tokens = word_tokenize(text)

print("\n2. Tokenized Words:")
print(tokens)

# Step 3: POS Tagging (Before Stopword Removal)
tagged_tokens = pos_tag(tokens)

print("\n3. POS Tagging:")
print(tagged_tokens)

# Step 4: Stopword Removal
stop_words = set(stopwords.words("english"))

filtered_tokens = [
    (word, tag) for word, tag in tagged_tokens
    if word.lower() not in stop_words and word.isalpha()
]

print("\n4. After Stopword Removal:")
print(filtered_tokens)

# Step 5: Stemming
stemmer = PorterStemmer()

stemmed_words = [
    stemmer.stem(word) for word, tag in filtered_tokens
]

print("\n5. Stemmed Words:")
print(stemmed_words)

# Step 6: Convert POS tags for WordNet
def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith("J"):
        return wordnet.ADJ
    elif treebank_tag.startswith("V"):
        return wordnet.VERB
    elif treebank_tag.startswith("N"):
        return wordnet.NOUN
    elif treebank_tag.startswith("R"):
        return wordnet.ADV
    else:
        return wordnet.NOUN

# Step 7: Lemmatization
lemmatizer = WordNetLemmatizer()

lemmatized_words = [
    lemmatizer.lemmatize(word, get_wordnet_pos(tag))
    for word, tag in filtered_tokens
]

print("\n6. Lemmatized Words:")
print(lemmatized_words)

print("\n" + "=" * 60)
print("PIPELINE COMPLETED")
print("=" * 60)
```
