```python
import pandas as pd
import networkx as nx

def run_lab7_facebook():
    print("Loading Facebook Dataset...")
    df = pd.read_csv("data/lab8_github_edges.csv")
    
    # Create Undirected Graph (Facebook Friendship Network)
    print("Building Undirected Graph...")
    G = nx.from_pandas_edgelist(
        df,
        'source',
        'target',
        create_using=nx.Graph()
    )

    print(f"Nodes: {G.number_of_nodes()} | Edges: {G.number_of_edges()}")
    
    # Apply PageRank
    print("Computing PageRank (alpha=0.85)...")
    pagerank = nx.pagerank(G, alpha=0.85)
    
    # Sort ranking
    sorted_pr = sorted(pagerank.items(), key=lambda x: x[1], reverse=True)

    # Display Top 10
    print("\n--- Top 10 Facebook Nodes by PageRank ---")
    for i, (page, rank) in enumerate(sorted_pr[:10]):
        print(f"Rank {i+1}: Node {page} = {rank:.6f}")

if __name__ == "__main__":
    run_lab7_facebook()
```
