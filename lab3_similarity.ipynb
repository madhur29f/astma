```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import pairwise_distances
import pandas as pd

# Documents
d0 = "This is a sample notebook"
d1 = "This notebook is sample"
d2 = "This is another sample notebook"

docs = [d0, d1, d2]

# ---------------- TF-IDF ----------------
print("=" * 60)
print("TF-IDF ANALYSIS")
print("=" * 60)

tfidf = TfidfVectorizer()
result = tfidf.fit_transform(docs)

print("\nIDF Values:")
for word, val in zip(tfidf.get_feature_names_out(), tfidf.idf_):
    print(f"{word:10} : {val:.4f}")

print("\nWord Indexes:")
print(tfidf.vocabulary_)

print("\nTF-IDF Matrix:")
print(result.toarray())

# ---------------- COUNT VECTOR ----------------
print("\n" + "=" * 60)
print("DOCUMENT TERM MATRIX")
print("=" * 60)

cv = CountVectorizer()
matrix = cv.fit_transform(docs)

dtm_df = pd.DataFrame(
    matrix.toarray(),
    columns=cv.get_feature_names_out(),
    index=["d0", "d1", "d2"]
)

print(dtm_df)

# ---------------- COSINE ----------------
print("\n" + "=" * 60)
print("COSINE SIMILARITY")
print("=" * 60)

cosine_sim = cosine_similarity(matrix)

cosine_df = pd.DataFrame(
    cosine_sim,
    index=["d0", "d1", "d2"],
    columns=["d0", "d1", "d2"]
)

print(cosine_df)

# ---------------- JACCARD ----------------
print("\n" + "=" * 60)
print("JACCARD SIMILARITY")
print("=" * 60)

binary_cv = CountVectorizer(binary=True, token_pattern=r"(?u)\b\w+\b")
X = binary_cv.fit_transform(docs)

jaccard_dist = pairwise_distances(X.toarray(), metric="jaccard")
jaccard_sim = 1 - jaccard_dist

jaccard_df = pd.DataFrame(
    jaccard_sim,
    index=["d0", "d1", "d2"],
    columns=["d0", "d1", "d2"]
)

print(jaccard_df)
```
