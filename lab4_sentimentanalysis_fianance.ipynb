```python
!pip install pandas
!pip install vaderSentiment
%pip install sklearn.metrics
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from sklearn.metrics import classification_report, accuracy_score

def analyze_finance_data():
    analyzer = SentimentIntensityAnalyzer()

    print("=" * 70)
    print("Loading Finance Dataset...")
    print("=" * 70)

    try:
        df = pd.read_csv("data/lab5_sentiment_finance.csv", encoding="latin-1").dropna()
    except FileNotFoundError:
        print("Error: lab5_sentiment_finance.csv not found.")
        return

    # Prediction Function
    def get_vader_label(text):
        score = analyzer.polarity_scores(str(text).lower())["compound"]

        if score >= 0.05:
            return "Positive"
        elif score <= -0.05:
            return "Negative"
        else:
            return "Neutral"

    print("Applying VADER Sentiment Analysis...")
    df["vader_prediction"] = df["text"].apply(get_vader_label)

    # Format true labels
    df["true_label"] = df["category"].str.capitalize()

    valid_df = df.dropna(subset=["true_label", "vader_prediction"])

    # Sample Predictions
    print("\n" + "=" * 70)
    print("Sample Predictions")
    print("=" * 70)

    for i, row in valid_df.head(5).iterrows():
        print(f"\nNews {i+1}:")
        print(f"Text       : {str(row['text'])[:120]}...")
        print(f"Predicted  : {row['vader_prediction']}")

    # Accuracy
    acc = accuracy_score(valid_df["true_label"], valid_df["vader_prediction"])

    print("\n" + "=" * 70)
    print("Model Performance")
    print("=" * 70)
    print(f"Accuracy : {acc:.4f}")

    print("\nClassification Report:")
    print(classification_report(valid_df["true_label"], valid_df["vader_prediction"]))

if __name__ == "__main__":
    analyze_finance_data()
```
